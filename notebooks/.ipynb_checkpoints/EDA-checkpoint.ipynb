- **we need to see important time series companents such as Trend, seasanality, error more clearly by decompoasition of the data.**

```python
#time series data decomposesion with aditive method with period about 24

def decomposision(feature):
    
    docompose_additive_model = seasonal_decompose(feature , model = "additive" , period = 24)
    trend_additive = docompose_additive_model.trend
    seasonality_additive = docompose_additive_model.seasonal
    residual_additive = docompose_additive_model.resid

    plt.figure(figsize = (12,9))
    plt.subplot(411)
    plt.plot(feature , label = "original")
    plt.legend()
    plt.subplot(412)
    plt.plot(trend_additive , label = "trend")
    plt.legend()
    plt.subplot(413)
    plt.plot(seasonality_additive , label = "seasonality")
    plt.legend()
    plt.subplot(414)
    plt.plot(residual_additive , label = "residual")
    plt.legend()
    plt.show()
    
```

```python
decomposision(weather_data["Temperature"])
```

- **becouse of the high number of cycles we see this like plot.**
- **Trend of the Temperature is increasing and decreasing, along with some spikes in between.**
- **there is no accurate seasionality in temperature and residual is high.**

```python
decomposision(weather_data["Relative_humidity"])
```

- **Trend have very high spikes and not accurate comparate to temperature.**
- **Seasionality and residuals are almost similar to temperature.**

#### Stationarity Test

##### 1) Rolling Mean & Standard Deviation Test

```python
plt.figure(figsize=(16,7))
fig = plt.figure(1)

# Plot rolling statistics:
orig = plt.plot(weather_data["Temperature"], color='blue', label='Original')
mean = plt.plot(weather_data["Rolling Mean of Temperature"], color='red', label='Rolling Mean')
std = plt.plot(weather_data["Rolling std of Temperature"], color='black', label='Rolling Std')

plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation')
plt.show(block=False)
```

- **Mean & Standard deviations are variating but data to be stationary when mean and std are always constant.**
- **this test conclude data(Temperature) is not a stationary data.**

##### 2) Augmented Dickey-Fuller Unit Root Test

```python
def stationarity_check(column_name):

    result = adfuller(column_name.dropna())

    print("ADF Statistic:", result[0])
    print("p-value:", result[1])
    print("Number of Lags Used:", result[2])
    print("Number of Observations Used:", result[3])

    print("\nCritical Values:")
    for key, value in result[4].items():
        print(f"{key}: {value}")

    if result[1] <= 0.05:
        print("The time series is Stationary (Reject the Null Hypothesis).")
    else:
        print("The time series is Non-Stationary (Fail to Reject the Null Hypothesis).")
```

```python
stationarity_check(weather_data["Temperature"])
```

```python
stationarity_check(weather_data["Relative_humidity"])
```

```python
stationarity_check(weather_data["Rain"])
```

```python
stationarity_check(weather_data["Soil_temperature"])
```

```python
stationarity_check(weather_data["Wind_speed"])
```

```python
stationarity_check(weather_data["Surface_pressure"])
```

```python
stationarity_check(weather_data["Pressure_msl"])
```

- **but this adf test shows each time series data is stationary.**

- we are not sure about one decision that is whether this data is stationary are not.
- So,we need to apply both stationary based models(ARMA,SARIMA) and non stationary based models(ML) as well.
