### making sweetviz Report

```python
os.chdir(ROOT_DIR/REPORTS)
start = time.time()
report = sv.analyze(lohore_dataset)
report.show_html("lohore_dataset_basic_report.html")
end = time.time()
print(f"TIme taken: {round(end-start,2)}sec")
cpu_percent = ps.cpu_percent(interval=1)
print(f"CPU Usage: {cpu_percent}%")
print("Report Downloaded Successfully!")
```
